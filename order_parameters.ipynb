```python
import numpy as np
import pandas as pd
import seaborn as sns
import h5py
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import pairwise_distances
```

```python
name = 'day1_fish19-barcodejoined'
f = h5py.File('/Users/viveksridhar/Dropbox/Network analysis of fish schools/fish_data/matched_tracks/' + name + '.h5', "r")
barcodejoined = True
```

```python
group = f['fields/']
```

```python
if (barcodejoined): 
    n_inds = len(f.attrs['actualids'])
else: 
    n_inds = 19

frame = (group['frame'].value).reshape((1,-1))
pos_x = (group['xcentroid'].value).reshape((1,-1))
pos_y = (group['ycentroid'].value).reshape((1,-1))
dir_x = (group['heading_x'].value).reshape((1,-1))
dir_y = (group['heading_y'].value).reshape((1,-1))

if (barcodejoined):
    identity = np.repeat(f.attrs['actualids'], frame.shape[1] / n_inds).reshape((1,-1))
else:
    identity = np.tile(range(0,n_inds), max(frame[0]) + 1).reshape((1,-1))
```

```python
data = [frame, identity, pos_x, pos_y, dir_x, dir_y]
data = np.array(data).T.reshape((dir_x.shape[1], len(data)))
df = pd.DataFrame(data, columns=['frame', 'ids', 'pos_x', 'pos_y', 'dir_x', 'dir_y'])
```

```python
f.close()
```

```python
# drop any inf in data
df = df.replace([np.inf, -np.inf], np.nan).dropna(axis=0, how="any")
```

```python
df_ord = df.groupby(['frame']).mean()
df_ord = df_ord.reset_index()

df_ord = df_ord.drop(columns=['ids'])
df_ord = df_ord.rename(index=str, columns={'pos_x':'centroid_x', 'pos_y':'centroid_y', 'dir_x':'gdir_x', 'dir_y':'gdir_y'})
```

```python
df = pd.merge(df, df_ord, how='left')
```

```python
df['centroid_disp_x'] = (df['pos_x'] - df['centroid_x'])
df['centroid_disp_y'] = (df['pos_y'] - df['centroid_y'])
df['centroid_disp_x'] = df['centroid_disp_x'] / np.linalg.norm((df['centroid_disp_x'], df['centroid_disp_y']), axis=0)
df['centroid_disp_y'] = df['centroid_disp_y'] / np.linalg.norm((df['centroid_disp_x'], df['centroid_disp_y']), axis=0)
```

```python
centroid_disp = [df['centroid_disp_x'], df['centroid_disp_y']]
direction = [df['dir_x'], df['dir_y']]
```

```python
df['rotation'] = np.cross(np.array(centroid_disp).T, np.array(direction).T)
```

```python
df_ord = df.groupby(['frame']).agg(['sum', 'count'])
df_ord = df_ord.reset_index()
```

```python
df_ord['polarisation'] = np.sqrt(df_ord['dir_x']['sum']**2 + df_ord['dir_y']['sum']**2) / df_ord['dir_x']['count']
df_ord['rotation'] = df_ord['rotation']['sum'].divide(df_ord['rotation']['count'])
```

```python
plt.figure(figsize=(30,3))
plt.plot(df_ord['frame'], np.abs(df_ord['rotation']['sum']), 'r', df_ord['frame'], df_ord['polarisation'], 'b')
plt.show()
```

```python
sns.jointplot(df_ord['polarisation'], np.abs(df_ord['rotation']['sum']), xlim=[0,1], ylim=[0,1], kind="hex", space=0, size = 7.5, stat_func=None, ratio=3)
plt.show()
```
